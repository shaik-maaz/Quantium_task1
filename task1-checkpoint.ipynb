```python

# Grouping and summing the total sales
grouped_chipsdata = chips_data.groupby(['PREMIUM_CUSTOMER', 'LIFESTAGE'], as_index=False)['TOT_SALES'].sum()

total_sales = grouped_chipsdata['TOT_SALES'].sum()

grouped_chipsdata['PERCENTAGE_SALES'] = (grouped_chipsdata['TOT_SALES'] / total_sales) * 100

heatmap_data = grouped_chipsdata.pivot(index='PREMIUM_CUSTOMER', columns='LIFESTAGE', values='PERCENTAGE_SALES')

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt='.2f', cmap='viridis', cbar=True)

plt.title("Customer Segment-wise Total Sales (%)")

plt.show()
```

```python
# Convert 'DATE' column to datetime
chips_data['DATE'] = pd.to_datetime(chips_data['DATE'])

# Filter the data for customer segment
budget_data = chips_data[chips_data['PREMIUM_CUSTOMER'] == 'Budget']
mainstream_data = chips_data[chips_data['PREMIUM_CUSTOMER'] == 'Mainstream']
premium_data = chips_data[chips_data['PREMIUM_CUSTOMER'] == 'Premium']

# Group by month and calculate total sales
monthly_sales_budget = budget_data.groupby(budget_data['DATE'].dt.to_period('M'))['TOT_SALES'].sum().reset_index()
monthly_sales_mainstream = mainstream_data.groupby(mainstream_data['DATE'].dt.to_period('M'))['TOT_SALES'].sum().reset_index()
monthly_sales_premium = premium_data.groupby(premium_data['DATE'].dt.to_period('M'))['TOT_SALES'].sum().reset_index()

# Plotting the result
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales_budget['DATE'].astype(str),monthly_sales_budget['TOT_SALES'],marker='o',color='blue',label='budget')
plt.plot(monthly_sales_mainstream['DATE'].astype(str),monthly_sales_mainstream['TOT_SALES'],marker='o',color='green',label ='mainstream')
plt.plot(monthly_sales_premium['DATE'].astype(str),monthly_sales_premium['TOT_SALES'],marker='o',color='red',label='premium')

# Adding title and labels
plt.title("Monthly Total Sales By All Customers Segments ", fontsize=16)
plt.xlabel("Months", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)

# Displaying the plot
plt.xticks(rotation=45)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

we can see that maximum sales are from budget and mainstream customers.Hence we concentrate on budget and mainstream customer purchasing products.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

chips_data['DATE'] = pd.to_datetime(chips_data['DATE'])

month_sales = chips_data.groupby(chips_data['DATE'].dt.to_period('M'))['TOT_SALES'].sum()

month_numbers = np.arange(len(month_sales))

slope, intercept = np.polyfit(month_numbers, month_sales.values, 1)


# Generate predictions y = Mx+C
predictions = slope * month_numbers + intercept

plt.plot(month_sales.index.to_timestamp().strftime('%b %Y'), month_sales.values, marker ='o', color='green', label='Monthly Sales')
plt.plot(month_sales.index.to_timestamp().strftime('%b %Y'), predictions, color='red', linestyle='--', label='Regression Line')

# Adding titles and labels
plt.title("Monthly Chip Sales with Linear Regression", fontsize=14)
plt.xlabel("Months")
plt.ylabel("Sales")

plt.xticks(rotation=45, ha='center')

plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

if slope > 0:
    print("Sales are increasing over time.")
elif slope < 0:
    print("Sales are decreasing over time.")
else:
    print("Sales are stable over time.")
```
